### Libraries <a id="part-one"></a>

```python
import matplotlib.pyplot as plt
```

### Code <a id="part-two"></a>

```python
flavors = ('Chocolate', 'Vanilla', 'Pistachio', 'Mango', 'Strawberry')
votes = (13, 11, 4, 8, 7)
colors = ('#8B4513', '#FFF8DC', '#93C572', '#E67F0D', '#D53032')
explode = (0.1, 0, 0, 0, 0)

plt.title('Favorite Ice Cream Flavors')
plt.pie(
    votes,
    labels=flavors,
    autopct='%1.1f%%',
    colors=colors,
    explode=explode,
    shadow=True
    )
plt.show()
```

### Explanation <a id="part-three"></a>

My code makes a simple pie chart. I made four different tuples, flavors, votes, colors and explode. Flavors and votes(people whoe like the flavor) is the data that makes up the pie chart. We pass plt.pie() the number of votes for each flavor and then the labels that we want to use which is the flavors. The colors tuple has five different html color codes for each flavor. Autopct value is a format string that shows the precison of the percentage shown. The explode tuple is what makes chocalate stand out since it has a value of 0.1 pushing it slightly out of the center. Shadow being set to True is what makes our pie chart look 3-D. Finally, plt.title allows us to give our pie chart a title above it.
